```python
import pandas as pd
```

```python
dataset_valid_smiles_file = "../SDF_to_SMILES/Dataset_valid_smiles.csv"
dataset_invalid_smiles_file = "../SDF_to_SMILES/Dataset_invalid_smiles.csv"
sdf_prop_file = "../SDF_SMILES_Prop/sdf_prop.csv"
dataset_valid_smiles_prop_file = "../SDF_SMILES_Prop/Dataset_valid_smiles_prop.csv"
dataset_invalid_smiles_prop_file = "../SDF_SMILES_Prop/Dataset_invalid_smiles_prop.csv"
dataset_smiles_prop_file = "../SDF_SMILES_Prop/Dataset_smiles_prop.csv"
```

```python
dataset_valid_smiles = pd.read_csv(dataset_valid_smiles_file)
dataset_invalid_smiles = pd.read_csv(dataset_invalid_smiles_file)
sdf_prop = pd.read_csv(sdf_prop_file)
```

```python
# sdf_prop.to_dict()["Filename"].items()
name2ind = {v: k for k, v in sdf_prop.to_dict()["Filename"].items()}
```

```python
dataset_valid_smiles.count()["name"]
```

```python
dataset_invalid_smiles.count()["name"]
```

```python
dataset_valid_smiles_prop = []
for i in range(dataset_valid_smiles.count()["name"]):
    name = dataset_valid_smiles.iloc[i,:]["name"][:-4]
    ind = name2ind[name]
    dataset_valid_smiles_prop.append({"name":name,"SMILES":dataset_valid_smiles.iloc[i,:]["SMILES"],"HOMO(eV)":sdf_prop.iloc[ind,:]["HOMO(eV)"],"LUMO(eV)":sdf_prop.iloc[ind,:]["LUMO(eV)"]})
dataset_valid_smiles_prop = pd.DataFrame(dataset_valid_smiles_prop)
dataset_valid_smiles_prop.to_csv(dataset_valid_smiles_prop_file, index=False)
```

```python
dataset_invalid_smiles_prop = []
for i in range(dataset_invalid_smiles.count()["name"]):
    name = dataset_invalid_smiles.iloc[i,:]["name"][:-4]
    ind = name2ind[name]
    dataset_invalid_smiles_prop.append({"name":name,"SMILES":dataset_invalid_smiles.iloc[i,:]["SMILES"],"HOMO(eV)":sdf_prop.iloc[ind,:]["HOMO(eV)"],"LUMO(eV)":sdf_prop.iloc[ind,:]["LUMO(eV)"],})
dataset_invalid_smiles_prop = pd.DataFrame(dataset_invalid_smiles_prop)
dataset_invalid_smiles_prop.to_csv(dataset_invalid_smiles_prop_file, index=False)
```

```python
pd.concat([dataset_valid_smiles_prop, dataset_invalid_smiles_prop], axis=0).to_csv(dataset_smiles_prop_file, index=False)
```
