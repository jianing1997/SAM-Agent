```python
sdf_smiles_prop_file = r"./SDF_SMILES_Prop/sdf_valid_smiles_prop.csv"
sdf_train_homo = r"./SDF_SMILES_Prop/sdf_train_homo.csv"
sdf_train_lumo = r"./SDF_SMILES_Prop/sdf_train_lumo.csv"
# sdf_smiles_prop_file = r"../SDF_SMILES_Prop/Dataset_valid_smiles_prop.csv"
# sdf_train_homo = r"../SDF_SMILES_Prop/Dataset_train_homo.csv"
# sdf_train_lumo = r"../SDF_SMILES_Prop/Dataset_train_lumo.csv"
```

```python
sdf_smiles_prop = pd.read_csv(sdf_smiles_prop_file)
```

```python
# {"SMILES":"smiles", "TARGET":"HOMO"}
train_data_homo = sdf_smiles_prop.loc[:,["SMILES","HOMO(eV)"]]
train_data_homo.columns = ["SMILES","TARGET"]
train_data_homo.to_csv(sdf_train_homo,index=False,columns=["SMILES","TARGET"])
```

```python
# {"SMILES":"smiles", "TARGET":"LUMO"}
train_data_lumo = sdf_smiles_prop.loc[:,["SMILES","LUMO(eV)"]]
train_data_lumo.columns = ["SMILES","TARGET"]
train_data_lumo.to_csv(sdf_train_lumo,index=False,columns=["SMILES","TARGET"])
```
