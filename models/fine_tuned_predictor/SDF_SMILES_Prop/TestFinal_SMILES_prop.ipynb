```python
import pandas as pd
```

```python
TestFinal_valid_smiles_file = "./SDF_to_SMILES/TestFinal_valid_smiles.csv"
TestFinal_invalid_smiles_file = "./SDF_to_SMILES/TestFinal_invalid_smiles.csv"
TestFinal_prop_file = "./SDF_SMILES_Prop/TestFinal_prop.csv"
TestFinal_valid_smiles_prop_file = "./SDF_SMILES_Prop/TestFinal_valid_smiles_prop.csv"
TestFinal_invalid_smiles_prop_file = "./SDF_SMILES_Prop/TestFinal_invalid_smiles_prop.csv"
TestFinal_smiles_prop_file = "./SDF_SMILES_Prop/TestFinal_smiles_prop.csv"
```

```python
TestFinal_valid_smiles = pd.read_csv(TestFinal_valid_smiles_file)
TestFinal_invalid_smiles = pd.read_csv(TestFinal_invalid_smiles_file)
TestFinal_prop = pd.read_csv(TestFinal_prop_file)
```

```python
name2ind = {v: k for k, v in TestFinal_prop.to_dict()["Filename"].items()}
```

```python
TestFinal_valid_smiles.count()["name"]
```

```python
TestFinal_invalid_smiles.count()["name"]
```

```python
TestFinal_valid_smiles_prop = []
for i in range(TestFinal_valid_smiles.count()["name"]):
    name = TestFinal_valid_smiles.iloc[i,:]["name"][:-4]
    ind = name2ind[name]
    TestFinal_valid_smiles_prop.append({"name":name,"SMILES":TestFinal_valid_smiles.iloc[i,:]["SMILES"],"HOMO(eV)":TestFinal_prop.iloc[ind,:]["HOMO(eV)"],"LUMO(eV)":TestFinal_prop.iloc[ind,:]["LUMO(eV)"]})
TestFinal_valid_smiles_prop = pd.DataFrame(TestFinal_valid_smiles_prop)
TestFinal_valid_smiles_prop.to_csv(TestFinal_valid_smiles_prop_file, index=False)
```

```python
TestFinal_invalid_smiles_prop = []
for i in range(TestFinal_invalid_smiles.count()["name"]):
    name = TestFinal_invalid_smiles.iloc[i,:]["name"][:-4]
    ind = name2ind[name]
    TestFinal_invalid_smiles_prop.append({"name":name,"SMILES":TestFinal_invalid_smiles.iloc[i,:]["SMILES"],"HOMO(eV)":TestFinal_prop.iloc[ind,:]["HOMO(eV)"],"LUMO(eV)":TestFinal_prop.iloc[ind,:]["LUMO(eV)"],})
TestFinal_invalid_smiles_prop = pd.DataFrame(TestFinal_invalid_smiles_prop)
TestFinal_invalid_smiles_prop.to_csv(TestFinal_invalid_smiles_prop_file, index=False)
```

```python
pd.concat([TestFinal_valid_smiles_prop,TestFinal_invalid_smiles_prop], axis=0).to_csv(TestFinal_smiles_prop_file, index=False)
```
