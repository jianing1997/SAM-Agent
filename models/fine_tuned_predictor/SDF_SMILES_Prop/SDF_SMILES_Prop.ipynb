```python
import pandas as pd
```

```python
sdf_valid_smiles_file = "./SDF_to_SMILES/sdf_valid_smiles.csv"
sdf_invalid_smiles_file = "./SDF_to_SMILES/sdf_invalid_smiles.csv"
sdf_prop_file = "./SDF_SMILES_Prop/sdf_prop.csv"
sdf_valid_smiles_prop_file = "./SDF_SMILES_Prop/sdf_valid_smiles_prop.csv"
sdf_invalid_smiles_prop_file = "./SDF_SMILES_Prop/sdf_invalid_smiles_prop.csv"
sdf_smiles_prop_file = "./SDF_SMILES_Prop/sdf_smiles_prop.csv"
```

```python
sdf_valid_smiles = pd.read_csv(sdf_valid_smiles_file)
sdf_invalid_smiles = pd.read_csv(sdf_invalid_smiles_file)
sdf_prop = pd.read_csv(sdf_prop_file)
```

```python
# sdf_prop.to_dict()["Filename"].items()
name2ind = {v: k for k, v in sdf_prop.to_dict()["Filename"].items()}
```

```python
sdf_valid_smiles.count()["name"]
```

```python
sdf_invalid_smiles.count()["name"]
```

```python
sdf_valid_smiles_prop = []
for i in range(sdf_valid_smiles.count()["name"]):
    name = sdf_valid_smiles.iloc[i,:]["name"][:-4]
    ind = name2ind[name]
    sdf_valid_smiles_prop.append({"name":name,"SMILES":sdf_valid_smiles.iloc[i,:]["SMILES"],"HOMO(eV)":sdf_prop.iloc[ind,:]["HOMO(eV)"],"LUMO(eV)":sdf_prop.iloc[ind,:]["LUMO(eV)"]})
sdf_valid_smiles_prop = pd.DataFrame(sdf_valid_smiles_prop)
sdf_valid_smiles_prop.to_csv(sdf_valid_smiles_prop_file, index=False)
```

```python
sdf_invalid_smiles_prop = []
for i in range(sdf_invalid_smiles.count()["name"]):
    name = sdf_invalid_smiles.iloc[i,:]["name"][:-4]
    ind = name2ind[name]
    sdf_invalid_smiles_prop.append({"name":name,"SMILES":sdf_invalid_smiles.iloc[i,:]["SMILES"],"HOMO(eV)":sdf_prop.iloc[ind,:]["HOMO(eV)"],"LUMO(eV)":sdf_prop.iloc[ind,:]["LUMO(eV)"],})
sdf_invalid_smiles_prop = pd.DataFrame(sdf_invalid_smiles_prop)
sdf_invalid_smiles_prop.to_csv(sdf_invalid_smiles_prop_file, index=False)
```

```python
pd.concat([sdf_valid_smiles_prop,sdf_invalid_smiles_prop], axis=0).to_csv(sdf_smiles_prop_file, index=False)
```
