```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# ========= 全局字体设置（全部 normal） =========
plt.rcParams.update({
    'font.family': 'Arial',
    'font.size': 28,
    'font.weight': 'normal',
    'axes.labelweight': 'normal',
    'axes.titleweight': 'normal',
    'xtick.labelsize': 28,
    'ytick.labelsize': 28
})

# ========= 创建 figure =========
fig = plt.figure(figsize=(8, 6), dpi=600)

# 🔴 关键：固定坐标轴区域（两个图必须完全一样）
ax = fig.add_axes([0.18, 0.18, 0.72, 0.72])
# [left, bottom, width, height]

# ========= 示例数据 =========
x = data["PVK_Eg"]
y = data["PCE"]

# 线性拟合
slope, intercept, _, _, _ = linregress(x, y)
fit_line_y = slope * x + intercept

# ========= 绘图 =========
ax.scatter(x, y, c='royalblue', alpha=0.7, s=80)
ax.plot(x, fit_line_y, color='coral', linewidth=2)

# ========= 轴标签 =========
ax.set_xlabel('Bandgap (eV)', fontweight='normal')
ax.set_ylabel('PCE (%)', fontweight='normal')

# ========= 刻度样式 =========
ax.tick_params(axis='both', which='major', width=2, length=6)

# ========= 固定边框线宽 =========
for spine in ax.spines.values():
    spine.set_linewidth(2)

plt.show()
```

```python
# 提取数据
x = data_PCE15["year"]
y = data_PCE15["PCE"]

# 执行最小二乘拟合
slope, intercept, _, _, _ = linregress(x, y)

# 创建空的DataFrame用于存放结果
df_year_high = pd.DataFrame(columns=data.columns)
df_year_low = pd.DataFrame(columns=data.columns)

# 将数据点根据相对于拟合直线的位置分为两组
for index, row in data_PCE15.iterrows():
    if row["PCE"] >= slope * row["year"] + intercept:
        df_year_high.loc[len(df_year_high)] = row
    else:
        df_year_low.loc[len(df_year_low)] = row

# 重置索引
df_year_high.reset_index(drop=True, inplace=True)
df_year_low.reset_index(drop=True, inplace=True)
print(df_year_low.shape)
print(df_year_high.shape)
df_year_low.head(3)
```

```python
df_year_high.head(3)
```

```python
# 创建空的 DataFrame 用于存放结果
df_high = pd.DataFrame(columns=df_Eg_high.columns)

# 循环遍历 df_Eg_high，检查每一行的 index_number 是否在 df_year_high 中出现
for _, row in df_Eg_high.iterrows():
    if row['index_number'] in df_year_high['index_number'].values:
        # 将符合条件的行添加到新的 DataFrame 中
        df_high.loc[len(df_high)] = row
print(df_high.shape)
df_high.head(3)
```

```python
# 创建空的 DataFrame 用于存放结果
df_low = pd.DataFrame(columns=df_Eg_low.columns)
for _, row in df_Eg_low.iterrows():
    if row['index_number'] in df_year_low['index_number'].values:
        # 将符合条件的行添加到新的 DataFrame 中
        df_low.loc[len(df_low)] = row
print(df_low.shape)
df_low.head(3)
```

```python
# 创建空的 DataFrame 用于存放结果
df_middle1 = pd.DataFrame(columns=df_Eg_low.columns)
for _, row in df_Eg_low.iterrows():
    if row['index_number'] in df_year_high['index_number'].values:
        # 将符合条件的行添加到新的 DataFrame 中
        df_middle1.loc[len(df_middle1)] = row
print(df_middle1.shape)
df_middle1.head(3)
```

```python
# 创建空的 DataFrame 用于存放结果
df_middle2 = pd.DataFrame(columns=df_Eg_low.columns)
for _, row in df_Eg_high.iterrows():
    if row['index_number'] in df_year_low['index_number'].values:
        # 将符合条件的行添加到新的 DataFrame 中
        df_middle2.loc[len(df_middle2)] = row
print(df_middle2.shape)
df_middle2.head(3)
```

```python
df_middle = pd.concat([df_middle1, df_middle2], axis=0).reset_index(drop=True)
print(df_middle.shape)
df_middle.head(3)
```

```python

import numpy as np

# 获取三组数据的 PCE 列
data_middle = df_middle['PCE']
data_high = df_high['PCE']
data_low = df_low['PCE']
bins = np.arange(15, 26, 0.1)
# 设置字体和字号
plt.rcParams.update({'font.family': 'Arial', 'font.size': 28, 'font.weight': 'bold'})

# 设置图形大小和边框磅值
plt.figure(figsize=(12, 6), dpi=600)
plt.rcParams['axes.linewidth'] = 3

# 绘制柱状图
plt.hist(data_low, bins=bins, color='b', alpha=0.5, label='Low PCE class')
plt.hist(data_middle, bins=bins, color='r', alpha=0.5, label='Middle PCE class')
plt.hist(data_high, bins=bins, color='g', alpha=0.5, label='High PCE class')

# 添加图例和标签
plt.legend()
plt.xlabel('PCE(%)', weight='bold')
plt.ylabel('Counts', weight='bold')

# 获取坐标轴对象
ax = plt.gca()

# 设置刻度线的粗细
ax.tick_params(axis='both', which='major', width=3, length=10)
# 添加图例
ax.legend(frameon=False)  # 去掉图例的边框
# 显示图形
plt.show()
```
